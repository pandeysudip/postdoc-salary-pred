```python
#getting min and max salary only from salary estimate column
df[['Min Salary', 'Max Salary']]=df['Salary Estimate'].str.split('(').str.get(0).str.split('-', expand=True)
```

```python
#removing dollar and k sign
df['Min Salary']= df['Min Salary'].str.replace('$', '').str.replace('K', '')
df['Max Salary']= df['Max Salary'].str.replace('$', '').str.replace('K', '')
```

```python
#converting into int
df['Min Salary']=df['Min Salary'].astype(int)
df['Max Salary']=df['Max Salary'].astype(int)
```

```python
#making new column average salary
df['avg_salary']=(df['Min Salary']+df['Max Salary'])/2
```

```python
df
```

```python
df['Company Name']
```

```python
#removing rating number from Company Name. Since if there is no rating in Rating column there is no rating number in Company Name column)
df['company']=df.apply(lambda x: x['Company Name'] if x['Rating']<0 else x['Company Name'][:-4], axis=1 )
```

```python
df['Location']
```

```python
#making state column
df['State']=df['Location'].str.split(',').str.get(1)
```

```python
#looking job for each state
df['State'].value_counts()
```

```python
#checking if job posting are in headquaters 
df['job_headquaters']=df.apply(lambda x: 1 if x['Location'] == x['Headquarters'] else 0, axis=1)
```

```python
#dropig headquaters and job_headquaters columns
df.drop(columns=['Headquarters', 'job_headquaters', 'Competitors'], inplace=True)
```

```python
#removing employee from the end
df['Size']=df['Size'].apply(lambda x: x.split('E')[0])
df['Size'].value_counts()
```

```python
#replacing unknown to -1
df['Size']=df['Size'].replace('Unknown','-1')
```

```python
df['Size'].value_counts()
```

```python
#looking for one company job description
df['Job Description'][0]
```

```python
#job description length
df['desc_length'] = df['Job Description'].apply(lambda x: len(x))
df['desc_length']
```

```python
df['Job Title'].value_counts()
```

```python
df['Job Title']=df['Job Title'].replace('POSTDOCTORAL ASSOCIATE','Postdoctoral Associate').replace('Post Doc Fellow','Postdoctoral Fellow')
```
